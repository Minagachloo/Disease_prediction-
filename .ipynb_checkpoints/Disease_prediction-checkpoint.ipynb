# Train dataset

```python
encoder = LabelEncoder()
train_data['diagnosis'] = encoder.fit_transform(train_data['prognosis'])
train_data['diagnosis'].unique()
```

```python
def train_model(x, y):

    L = LogisticRegression()
    dt = DecisionTreeClassifier()
    rf = RandomForestClassifier(n_estimators=7)
    GB = GradientBoostingClassifier()
    kn = KNeighborsClassifier()
    svc = SVC()

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

    modles = [L, dt, rf, GB, kn, svc]
    modle_names = ['LogisticRegression', 'DecisionTreeClassifier', 'RandomForestClassifier', 'GradientBoostingClassifier', 'KNeighborsClassifier', 'SVC']

    result_list = []

    for model, model_name in zip(modles, modle_names):
        model.fit(x_train, y_train)
        p = model.predict(x_test)
        train_accuracy = accuracy_score(y_train, model.predict(x_train))
        test_accuracy = accuracy_score(y_test, p)

        result_list.append({'Model': model_name,
                            'Training Accuracy': train_accuracy,
                            'Validation Accuracy': test_accuracy
                            })

    result_df = pd.DataFrame(result_list)

    return result_df
```

```python
X = train_data.drop(['prognosis', 'diagnosis'], axis=1).values
y = train_data['diagnosis'].values

result_df = train_model(X, y)
```
